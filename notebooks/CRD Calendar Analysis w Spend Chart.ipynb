```python
# Import in Pandas functions
import pandas as pd
%matplotlib inline
from datetime import datetime
import matplotlib.pyplot as plt
from matplotlib import style
import matplotlib.dates as dt
import numpy as np

# The path to our CSV file
file = "Resources/CalendarData.csv"

# Read our AirBnB Calendar data into pandas
CalendarData = pd.read_csv(file)
CalendarData.head()
```

```python

#Identify the number of lines in CalendarData DataFrame.
CalendarData.shape
```

```python
# Used the next line to work with 100,000 lines of data as opposed to the population.
CalendarData = CalendarData[0:100000]
```

```python
# Eliminate "$" from 'price' column.
CalendarData['price'] = CalendarData['price'].str.replace('$','')
```

```python
# Eliminate "," from 'price' column.
CalendarData['price'] = CalendarData['price'].str.replace(',','')
```

```python
# Identify make-up of CalendarData and display the tail of Panadas.
CalendarData.tail()
```

```python
#Review of the CalendarData file after the "$" and "," are removed to change from string to float.
CalendarData.info
```

```python
CalendarData.columns = ['Listing ID', 'Date', 'Available', 'Price']
CalendarData.tail(20)
```

```python
# Drop all rows with no price data.
CalendarData = CalendarData.dropna()
```

```python

CalendarData.loc[:,'Price'] = CalendarData.loc[:,'Price'].astype('float')
```

```python
CalendarData.loc[:,'Date'] = CalendarData.loc[:,'Date'].astype('datetime64[ns]')
```

```python
CalendarData.dtypes
```

```python
CalendarData.to_csv("Output/CalendarDatabyDate.csv")
```

```python
DateGrouped_df = CalendarData.groupby(['Date'])
DateGrouped_df.head()
```

```python
# Identify the top Dates with number of purchases, average purchases, and total purchase value
Date_Spend_sum = CalendarData.groupby('Date')['Price'].sum()
Date_Spend_mean = CalendarData.groupby('Date')['Price'].mean()
Date_Spend_count = CalendarData.groupby('Date')['Price'].count()

# Sort data on date
Date_Spend = pd.DataFrame({'Total Spend by Date':Date_Spend_sum,'Average Spend by Date':Date_Spend_mean,'Number of Listings by Date':Date_Spend_count})

# Print output
Date_Spend.sort_values('Total Spend by Date',ascending=False)
```
