```python
# Import in Pandas functions
import pandas as pd
import datetime
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np

# The path to our CSV file
file = "Resources/CalendarData.csv"

# Read our AirBnB Calendar data into pandas
CalendarData = pd.read_csv(file)
CalendarData.head()
```

```python

#Identify the number of lines in CalendarData DataFrame.
CalendarData.shape
```

```python
# Used the next line to work with 10,000 lines of data as opposed to the population.
# CalendarData = CalendarData[0:10000]
```

```python
# Eliminate "$" from 'price' column.
CalendarData['price'] = CalendarData['price'].str.replace('$','')
```

```python
# Eliminate "," from 'price' column.
CalendarData['price'] = CalendarData['price'].str.replace(',','')
```

```python
# Identify make-up of CalendarData and display the tail of Panadas.
CalendarData.tail()
```

```python
#Review of the CalendarData file after the "$" and "," are removed to change from string to float.
CalendarData.info
```

```python
CalendarData.columns = ['Listing ID', 'Date', 'Available', 'Price']
CalendarData.tail(20)
```

```python
# Drop all rows with no price data.
CalendarData = CalendarData.dropna()
```

```python

CalendarData['Price'] = CalendarData['Price'].astype('float')
```

```python
CalendarData.dtypes
```

```python
CalendarData.to_csv("Output/CalendarDatabyDate.csv")
```

```python
DateGrouped_df = CalendarData.groupby(['Date'])
DateGrouped_df.head()
```

```python
# Identify the top Dates with number of purchases, average purchases, and total purchase value
Date_Spend_sum = CalendarData.groupby('Date')['Price'].sum()
Date_Spend_mean = CalendarData.groupby('Date')['Price'].mean()
Date_Spend_count = CalendarData.groupby('Date')['Price'].count()

# Sort data on date
Date_Spend = pd.DataFrame({'Total Spend by Date':Date_Spend_sum,'Average Spend by Date':Date_Spend_mean,'Number or Listings by Date':Date_Spend_count})

# Print output
Date_Spend.sort_values('Total Spend by Date',ascending=False)
```

```python
# Input the data about the holidays using years to represent each day of the holiday weekend for funtionality
Rental_Data = {'TG 2017': {2016: 2653192, 2017: 2745982, 2018: 2909169, 2019: 2958551}, 
                'Christmas 2017': {2016: 3276775, 2017: 3208079, 2018: 3129396, 2019: 3061068},
                'NYE 2018': {2016: 2871282, 2017: 2894549, 2018: 2683243, 2019: 0}, 
                'Memorial 2018': {2016: 2612120, 2017: 2522906, 2018: 2518031, 2019: 0},
                'July 4th 2018': {2016: 2529429, 2017: 2531422, 2018: 2532631, 2019: 0},
                'Labor 2018': {2016: 2673235, 2017: 2745982, 2018: 2909169, 2019: 0}}
# Set key for year and holiday data
year_data = {}
holidays = []

# Create for loop to cycle through data and append new data to create bar chart
for key, holiday_dict in Rental_Data.items():
    holidays.append(key)
    for year, revenue in sorted(holiday_dict.items()): 
        if year not in year_data:
            year_data[year] = []
        year_data[year].append(revenue)

# Use year data sorted in a logical way
years = sorted(year_data.keys())
year_sum = [0]*len(holidays)
bar_graphs = []

# Use year data to loop through data and reserve the year information
for year in years:
    graph = plt.bar(holidays, year_data[year], bottom=year_sum)
    bar_graphs.append(graph[0])
    year_sum = [year_sum[i] + year_data[year][i] for i in range(len(holidays))]
    
# Rotate the x-axis lables for readability    
plt.xticks(rotation=-30) 
# Creating title for y-axis
plt.ylabel('Revenue')
# Creating title for the plot
plt.title('Revenue During 2017 - 2018 Holidays')
# Creating legend for the plot
plt.legend(('Day 1', 'Day 2', 'Day 3', 'Day 4'), loc = 1, frameon = 'false')
# Store the plot
plt.savefig("Output/Revenue_for_Holidays.png")
# Show the plot
plt.show()
```
